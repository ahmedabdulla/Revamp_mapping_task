```python
import pandas as pd
old_maping = pd.read_csv('old_map.csv')
```

```python
# Creating a new column called new_name.

old_maping['new_name'] = old_maping['JOB_NAME'].apply(lambda x: x[15:]if x[:15] == 'm_STG_DWH_LOAD_'
                                                      else (x[11:] if x[:11] == 'm_DWH_LOAD_' else x))
```

```python
# I have add a new column called 'USED_TOOL' so that import the new maping with value USED_TOOL = 'INFORMATICA'
old_maping = old_maping[old_maping.USED_TOOL == 'SSIS']
old_maping.head()
```

```python
# We have 503 row in the old mapping.
old_maping.count()
```

# Informatica log for STG Maps. (To get the STG Tables)

```python
info_log_STG = pd.read_csv('info_STG_MAP.csv')
info_log_STG
```

```python
# I am intersted in teh column SESSION_NAME so I will extract it.
info_log_STG = info_log_STG.iloc[: , [3]]
info_log_STG
```

```python
# There is 252 STG map on the informatica log.
info_log_STG.count()
```

```python
# Adding a new column called 'new_name'
info_log_STG['new_name'] = info_log_STG['SESSION_NAME'].apply(lambda x: x[12:] if x[:12] == 's_m_STG_TRM_' 
                                                      else x)
```

```python
info_log_STG
```

# Informatica log for EWD Maps. (To get the EDW Tables)

```python
info_log_EWH = pd.read_csv('Target.csv')
```

```python
# Adding a new column called 'new_name'
info_log_EWH['new_name'] = info_log_EWH['SESSION_NAME'].apply(lambda x: x[8:] if x[:8] == 's_m_STG_' 
                                                      else x)
```

```python
# I don't need the GSTG Maps
info_log_EWH = info_log_EWH[info_log_EWH["SESSION_NAME"].str[:8] == 's_m_DWH_']
info_log_EWH = info_log_EWH[info_log_EWH["SESSION_NAME"].str[:12] !='s_m_DWH_GSTC']
info_log_EWH
```

```python
# Get the EDW tables names.
info_log_EWH['new_name'] = info_log_EWH['SESSION_NAME'].apply(lambda x: x[4:] if x[:4] == 's_m_' 
                                                      else x)
info_log_EWH
```

# joining the DataFrames (Old_mapping with staging maps in informatica)

```python
left_join = pd.merge(old_maping[old_maping["WORK_STRAM"] == 'STG'], info_log_STG, on='new_name', how='left')
left_join
```

```python
left_join.count()
```

# joining the DataFrames (Old_mapping with EDW maps in informatica)

```python
left_join_T = pd.merge(old_maping[old_maping["WORK_STRAM"] == 'DWH'], info_log_EWH, on='new_name', how='left')
left_join_T
```

```python
left_join_T.count()
```

```python
# Combine the two dataframes (STG Tables with EDW Tables)

com = pd.concat([left_join, left_join_T])
com['USED_TOOL'] = 'INFORMATICA'
com
```

```python
# There is Tables the not found in the revamp maps as the following
Col_not_match = com[pd.isnull(com["SESSION_NAME"])]
Col_not_match
```
